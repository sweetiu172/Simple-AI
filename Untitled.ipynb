```python
#tìm đường đi

def find_path(state):
    #tìm vị trí của số 0
    row = np.squeeze(np.where(state == 0)[0]) 
    col = np.squeeze(np.where(state == 0)[1])
    pos = []
    # up
    if (row - 1) >= 0:
        temp = state.copy()
        temp[row,col], temp[row-1,col] = temp[row-1,col], temp[row,col] #đổi vị trí
        pos.append(temp)
    #down
    if (row + 1) < len(state):
        temp = state.copy()
        temp[row,col], temp[row+1,col] = temp[row+1,col], temp[row,col] #đổi vị trí
        pos.append(temp)
    #right
    if (col + 1) < len(state):
        temp = state.copy()
        temp[row,col], temp[row,col+1] = temp[row,col+1], temp[row,col] #đổi vị trí
        pos.append(temp)
    #left
    if (col - 1) >= 0:
        temp = state.copy()
        temp[row,col], temp[row,col-1] = temp[row,col-1], temp[row,col] #đổi vị trí
        pos.append(temp)
    return pos
```

```python
#tính h-score
def h(state,goal):
    score = 0
    for i in range(len(state)):
        for j in range(len(state[i])):
            if state[i,j] != goal[i,j] and state[i,j] != 0:
                score += 1
    return score
```

```python
def checkInQueue(state, global_queue):
    if state.tolist() in global_queue:
        return True
    return False
```

```python
from math import inf
global_queue = []

state = initial_state.copy()
global_queue.append(state.tolist())

g_score = 0

print('*'*60)
print("Level: ", g_score)
print(state)
print('*'*60)

while np.array_equal(state, final_state) == False:
    
    g_score += 1
    possible_moves = find_path(state)

    f_score = inf
    for idx, move in enumerate(possible_moves):
        if checkInQueue(move, global_queue):
            continue
        temp_f_score = h(move,final_state) + g_score
        if temp_f_score < f_score:
            f_score = temp_f_score
            move_idx = idx
        state = possible_moves[move_idx].copy()
    global_queue.append(state.tolist())
    print('*'*60)
    print("Level: ", g_score)
    print(state)
    print('*'*60)
    
    if(g_score == 50):
        break
```
